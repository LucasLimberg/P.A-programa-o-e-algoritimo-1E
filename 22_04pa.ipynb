```python
#questao 1
#par ou impar

numero = int(input("digite um numero:  "))

if numero % 2 == 0:
  print(f"o numero {numero} e um par")

else:
  print(f"seu numero é impar")
```

```python
#questao 2
#maior de dois numeros

a = int(input("digite o primeiro numero:  "))
b = int(input("digite o primeiro numero:  "))

if a > b:
  print(f"{a} é maior que {b}")

elif a == b:
  print(f"{a} é igual a {b}")

else:
  print(f" {a} é menor que {b}")

```
