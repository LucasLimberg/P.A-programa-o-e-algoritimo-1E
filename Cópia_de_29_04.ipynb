```python
# escrver um programa que verifica se é uma palindroma

palavra = str(input("escreva uma palavra:  "))
invertida = ""

for i in (palavra):
  invertida = i + invertida

if palavra == invertida:
  print("sua palavra é uma palindroma")

else:
  print("sua palavra não é uma palindroma")
```
