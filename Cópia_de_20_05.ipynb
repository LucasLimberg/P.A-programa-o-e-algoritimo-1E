```python
#escrever um programa que recebe 04 notas do aluno, calcula a media e imprimi aprovado, recuperaçao ou reprovado

#regras

#utlizar while
#nota maior ou igual a 7 aprovado
# nota maior ou igual a 5 recuperaçao
#mota menor que 5 aprovado

notas = 0
bimestre = 1
soma = 0

try:

  while bimestre <= 4:
    notas = float(input(f"digita sua nota do {bimestre} bimestre:  "))

    bimestre += 1
    soma = soma + notas

    print(soma)

  if (soma / 4) >= 7:
    print(f" sua nota foi {soma/4} vc foi aprovado")

  elif (soma / 4) >= 5:
    print(f" sua nota foi {soma/4} vc foi reprovado")
  else:
    print(f" sua nota foi {soma/4} vc foi reprovado")

except:
  print ("digite apenas numeros")
```
