```python
#1 contagem simples,o usuario digita um numero e o programa imprimi do zero ate o numero

numero = int(input("digite um numero:  "))
for i in range(numero + 1):
  print (i)
```

```python
#2 contagem regressiva,o usuario digita um numero e o programa imprimi do numero ate o zero
numero = int(input("digite um numero:  "))
for i in range (numero+1):
  print (numero-i)


```

```python
#3 numeros pares,o usuario digita um numero e o programa imprimi
numero = int(input("digite um numero:  "))
for i in range (numero+1):
  if i %2 == 0:
    print (i)
```

```python
#4 numeros impares, o usuario digita um numwro e o programa imprimi os numeros impares

numero = int(input("digite um numero:  "))
for i in range (numero+1):
  if i %2 == 1:
    print (i)
```

```python
#5 soma,o usuario digita um numero e o programa soma os numeros do zero ate o numero
numero = int(input("digite um numero:  "))
soma = 0
for i in range (numero+1):
  soma += i

print(soma)
```

```python
#6 tabuada,o usuario digita um numero e o programa imprimi a tabuada do numero

numero = int(input("digite um numero:"))

for i in range(11):
  print(f"{i} x {numero} = {numero * i}")
```
