```python
#questao 2    Solicite ao usuário um número inteiro positivo.

#Utilizando um laço de reprodução, calcule e exiba a soma de
#todos os números de 1 até o número informado.

#Exemplo:

#Digite um número: 5

#Resultado: 15

#Pois:

#1 + 2 + 3 + 4 + 5 = 15
numero = int(input("digite um numero:  "))
soma = 0
for i in range (numero+1):
  soma += i

print(soma)

```
