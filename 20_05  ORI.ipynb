```python
# Você foi contratado para desenvolver um sistema simples de caixa eletrônico.

# O sistema deverá funcionar através de um menu interativo utilizando a estrutura de repetição while.

# O programa deve iniciar com um saldo inicial de R$ 1.000,00 e permitir que o usuário escolha diferentes operações até decidir sair do sistema.

# Funcionalidades:

# 1 - ver saldo;
# 2 - Depositar dinheiro
# 3 - sacar dinheiro
# 4 - sair

# Regras:

# ver saldo
# Ao escolher essa opção, o sistema deverá mostrar o saldo atual da conta;

# Depositar dinheiro
# O usuário deverá informar um valor para depósito
# O valor digitado deve ser somado ao saldo.

# Sacar dinheiro
# O usuário deverá informar o valor do saque.

# O sistema devera verificar:
# Se houver suficiente - realizar o saque.
# se nao houver saldo suficiente - mostrar mensgaem de "saldo insuficiente"

#sair

# o programa devera encerrar apenas quando o usuario escolher a opcao

saldo = 1000.00
opcao = 0

print("Bem-vindo ao Caixa Eletrônico!")

while opcao != 4:
  print("\n--- Menu --- ")
  print("1 - Ver Saldo")
  print("2 - Depositar Dinheiro")
  print("3 - Sacar Dinheiro")
  print("4 - Sair")
  print("------------")

  try:
    opcao = int(input("Digite a opção desejada: "))

    if opcao == 1:
      print(f"Seu saldo atual é de R$ {saldo:.2f}")
    elif opcao == 2:
      valor_deposito = float(input("Digite o valor a ser depositado: "))
      if valor_deposito > 0:
        saldo += valor_deposito
        print(f"Depósito de R$ {valor_deposito:.2f} realizado com sucesso. Novo saldo: R$ {saldo:.2f}")
      else:
        print("Valor de depósito inválido. Digite um valor positivo.")
    elif opcao == 3:
      valor_saque = float(input("Digite o valor a ser sacado: "))
      if valor_saque > 0:
        if saldo >= valor_saque:
          saldo -= valor_saque
          print(f"Saque de R$ {valor_saque:.2f} realizado com sucesso. Novo saldo: R$ {saldo:.2f}")
        else:
          print("Saldo insuficiente para realizar o saque.")
      else:
        print("Valor de saque inválido. Digite um valor positivo.")
    elif opcao == 4:
      print("Obrigado por usar o Caixa Eletrônico! Volte sempre.")
    else:
      print("Opção inválida. Por favor, escolha uma opção de 1 a 4.")
  except ValueError:
    print("Entrada inválida. Por favor, digite um número para a opção e para os valores.")
  except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
```
