```python
# Sistema de Votação

votos_ana = 0
votos_carlos = 0
votos_julia = 0
votos_brancos = 0
votos_nulos = 0

total_votos = 0

while total_votos < 6:

    print("\n====== CENTRAL DE VOTAÇÃO ======")
    print("1 - Ana Souza")
    print("2 - Carlos Silva")
    print("3 - Júlia Santos")
    print("--------------------------------")
    print("Digite 1, 2 ou 3 para votar.")
    print("ENTER para voto em branco.")
    print("Qualquer outro valor será considerado nulo.")

    escolha = input("Seu voto: ")

    if escolha == "1":
        votos_ana += 1
    elif escolha == "2":
        votos_carlos += 1
    elif escolha == "3":
        votos_julia += 1
    elif escolha == "":
        votos_brancos += 1
    else:
        votos_nulos += 1

    total_votos = (
        votos_ana +
        votos_carlos +
        votos_julia +
        votos_brancos
    )

    print("\nParcial da eleição:")
    print(f"Ana Souza: {votos_ana}")
    print(f"Carlos Silva: {votos_carlos}")
    print(f"Júlia Santos: {votos_julia}")
    print(f"Brancos: {votos_brancos}")
    print(f"Nulos: {votos_nulos}")

print("\n===== RESULTADO FINAL =====")

votos_validos = votos_ana + votos_carlos + votos_julia

if votos_validos == 0:
    print("Nenhum candidato recebeu voto.")
elif votos_ana > votos_carlos and votos_ana > votos_julia:
    print(f"Ana Souza venceu com {votos_ana} votos.")
elif votos_carlos > votos_ana and votos_carlos > votos_julia:
    print(f"Carlos Silva venceu com {votos_carlos} votos.")
elif votos_julia > votos_ana and votos_julia > votos_carlos:
    print(f"Júlia Santos venceu com {votos_julia} votos.")
else:
    print("Eleição empatada.")
```
